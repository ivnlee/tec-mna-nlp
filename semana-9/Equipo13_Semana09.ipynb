**b.** Utiliza ahora un ajuste de parámetros (fine-tuning) con el modelo Transformer DistilBERT,
para generar un nuevo modelo y clasificación de los comentarios. Es decir, el objetivo es
utilizar un modelo Transformer pre-entrenado y adaptarlo (fine-tuning) o ajustarlo a las
características de tus datos de entrenamiento. Indicar desempeño (accuracy) obtenido.

```python
import torch
from transformers import AutoTokenizer, DistilBertForSequenceClassification
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import accuracy_score

# Cargar los conjuntos de datos de entrenamiento y validacion
train_comments = train_data["review"].tolist()  # Obtener los comentarios de entrenamiento como una lista
train_labels = train_data["rating"].tolist()  # Obtener las etiquetas de entrenamiento como una lista
test_comments = train_val_data["review"].tolist()  # Obtener los comentarios de validación como una lista
test_labels = train_val_data["rating"].tolist()  # Obtener las etiquetas de validación como una lista

# Tokenizar los comentarios utilizando el tokenizer de DistilBERT
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")  # Cargar el tokenizer de DistilBERT
train_encodings = tokenizer(train_comments, truncation=True, padding=True, max_length=512)  # Tokenizar los comentarios de entrenamiento
test_encodings = tokenizer(test_comments, truncation=True, padding=True, max_length=512)  # Tokenizar los comentarios de validación

# Crear los conjuntos de datos de entrenamiento y validacion
train_dataset = TensorDataset(torch.tensor(train_encodings.input_ids),
                              torch.tensor(train_encodings.attention_mask),
                              torch.tensor(train_labels))  # Crear un conjunto de datos para el entrenamiento
test_dataset = TensorDataset(torch.tensor(test_encodings.input_ids),
                             torch.tensor(test_encodings.attention_mask),
                             torch.tensor(test_labels))  # Crear un conjunto de datos para la validación

# Cargar el modelo pre-entrenado de DistilBERT
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)  # Cargar el modelo pre-entrenado de DistilBERT con 2 etiquetas

# Entrenamiento del modelo
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Verificar si hay una GPU disponible
model.to(device)  # Mover el modelo a la GPU si está disponible, de lo contrario, se utilizará la CPU
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)  # Cargar el conjunto de datos de entrenamiento en lotes
optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)  # Definir el optimizador Adam con una tasa de aprendizaje de 1e-5
loss_fn = torch.nn.CrossEntropyLoss()  # Definir la función de pérdida de entropía cruzada

model.train()  # Establecer el modo de entrenamiento del modelo
for epoch in range(3):  # Ejecutar el entrenamiento durante 3 épocas
    total_loss = 0
    for batch in train_loader:
        batch = tuple(t.to(device) for t in batch)
        input_ids, attention_mask, labels = batch

        optimizer.zero_grad()  # Reiniciar los gradientes acumulados en cada iteración
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)  # Obtener las salidas del modelo y calcular la pérdida
        loss = outputs.loss  # Obtener la pérdida
        total_loss += loss.item()  # Acumular la pérdida total

        loss.backward()  # Realizar la retropropagación para calcular los gradientes
        optimizer.step()  # Actualizar los parámetros del modelo utilizando el optimizador

    avg_loss = total_loss / len(train_loader)  # Calcular la pérdida promedio
    print(f"Epoch {epoch+1}/{3}, Loss: {avg_loss}")  # Imprimir la pérdida promedio de la época actual

# Evaluación del modelo
model.eval()  # Establecer el modo de evaluación del modelo
test_loader = DataLoader(test_dataset, batch_size=8)  # Cargar el conjunto de datos de validación en lotes
predicted_labels = []  # Lista para almacenar las etiquetas predichas
true_labels = []  # Lista para almacenar las etiquetas verdaderas

with torch.no_grad():
    for batch in test_loader:
        batch = tuple(t.to(device) for t in batch)
        input_ids, attention_mask, labels = batch

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)  # Obtener las salidas del modelo sin calcular la pérdida
        logits = outputs.logits  # Obtener los logits (valores de salida antes de aplicar la función softmax)
        predicted_labels.extend(torch.argmax(logits, dim=1).cpu().tolist())  # Obtener las etiquetas predichas y agregarlas a la lista
        true_labels.extend(labels.cpu().tolist())  # Obtener las etiquetas verdaderas y agregarlas a la lista

accuracy = accuracy_score(true_labels, predicted_labels)  # Calcular la precisión del modelo comparando las etiquetas verdaderas y predichas
print("Accuracy:", accuracy)  # Imprimir la precisión del modelo
```

```python
from sklearn.metrics import accuracy_score
from transformers import pipeline
from transformers import BertForSequenceClassification


# Cargamos el modelo y el tokenizer de BERT en español
model_name = "dccuchile/bert-base-spanish-wwm-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# Comentarios en español
text_list = translated_comments["reviews"].tolist()# Lista de comentarios en español
label_list = translated_comments["labels"].tolist()# Lista de etiquetas de sentimiento reales (0 para negativo, 1 para positivo)
predicted_labels = []

# Usamos el modelo y tokenizer para hacer el análisis de sentimientos cada comentario
for text, label in zip(text_list, label_list):
    encoded_input = tokenizer(text, padding=True, truncation=True, return_tensors="pt")
    logits = model(**encoded_input).logits
    predicted_class = torch.argmax(logits, dim=1).item()
    predicted_labels.append(predicted_class)

# Calculamos la precisión que obtuvimos
accuracy = accuracy_score(label_list, predicted_labels)
print(f"Accuracy: {accuracy}")
```

Como se observa, se está acertando el sentimiento en el 62% de los comentarios, para intentar mejorar el acierto, vamos a realizar algunos pasos más de pre-procesamiento.

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# Vamos a utilizar stemmer y obtener las stopwords en español
stemmer = SnowballStemmer("spanish")
stopwords_es = set(stopwords.words("spanish"))

# Creamos una función para aplicar el preprocesamiento
def preprocess_text(text):
    # se convierte todo a minúsculas
    text = text.lower()
    # removemos los caracteres especiales y puntuación
    text = re.sub(r"[^a-zA-Z0-9áéíóúüñ]", " ", text)
    # Tokenizamos
    tokens = text.split()
    # Se eliminan las stopwords
    tokens = [token for token in tokens if token not in stopwords_es]
    # Aplicamos stemming
    tokens = [stemmer.stem(token) for token in tokens]
    # unimos los tokens nuevamente
    preprocessed_text = " ".join(tokens)
    return preprocessed_text

# Generamos las listas con las etiquetas observadas y el nuevo conjunto ya preprocesado
label_list = translated_comments["labels"].tolist()
preprocessed_text_list = [preprocess_text(text) for text in translated_comments["reviews"].tolist()]

predicted_labels = []

# Usamos el modelo y tokenizer para hacer el análisis de sentimientos cada comentario
for text in preprocessed_text_list:
    encoded_input = tokenizer(text, padding=True, truncation=True, return_tensors="pt")
    logits = model(**encoded_input).logits
    predicted_class = torch.argmax(logits, dim=1).item()
    predicted_labels.append(predicted_class)

# Calculamos la precisión que obtuvimos
accuracy = accuracy_score(label_list, predicted_labels)
print(f"Accuracy: {accuracy}")
```

Con el preprocesamiento de los datos obtuvimos una mejoría importante, ahora el modelo está acertando el sentimiento del 75% de los comentarios.
